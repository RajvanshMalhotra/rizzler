```python
import json


# with open('your_file.json', 'r') as f:
#     dataset = json.load(f)

def transform_to_dynamic_llama(data):
    mappings = {
        "universal": ("Poetry", "Sincerity"),
        "old_school_romantic": ("Aviation", "Nostalgia"),
        "confident": ("Fine Dining", "Boldness"),
        "genuine": ("Real Talk", "Simplicity"),
        "romantic": ("Theme Parks", "Happiness"),
        "playful": ("Social Situations", "Humor"),
        "clever": ("Modern Trends", "Witty References"),
        "bold": ("Confidence", "Directness"),
        "direct": ("Raw Honesty", "Clarity"),
        "ted_mosby": ("Destiny", "Life Stories"),
        "jim_halpert": ("Workplace Life", "Disarming Honesty"),
        "shakespeare": ("Classical Literature", "Eternal Devotion"),
        "joey_tribbiani": ("Casual Connection", "Charm"),
        "chandler_bing": ("Self-Deprecation", "Vulnerable Humor"),
        "ross_geller": ("Science/History", "Academic Interest"),
        "leonard_hofstadter": ("Physics", "Alternative Realities"),
        "sheldon_cooper": ("Logic", "Analytical Proof"),
        "raj_koothrappali": ("Cinema", "Romantic Gestures"),
        "howard_wolowitz": ("Engineering", "Technical Precision"),
        "phil_dunphy": ("Real Estate", "Optimism"),
        "claire_dunphy": ("Organization", "Spontaneity"),
        "gloria_delgado": ("Heritage", "Fierce Intensity"),
        "jay_pritchett": ("Experience", "Straight Talk"),
        "cameron_tucker": ("The Arts", "Theatricality")
    }

    formatted_rows = []
    for entry in data:
        char = entry.get("character_energy", "universal")
        x, y = mappings.get(char, ("Life", "Honesty"))

        row = {
            "instruction": "Generate a high-conviction romantic line for a person with specific interests.",
            "input": f"Target: Into [X: {x}] and loves [Y: {y}].",
            "output": entry["line"]
        }
        formatted_rows.append(row)
    return formatted_rows

```

```python
from trl import SFTConfig, SFTTrainer
from datasets import Dataset


dataset = Dataset.from_list(result)

alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token # Must add EOS_TOKEN

def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # Must add EOS_TOKEN, otherwise generation will go on forever!
        text = alpaca_prompt.format(instruction, input, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }

dataset = dataset.map(formatting_prompts_func, batched = True)

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    packing = False, # Can make training 5x faster for short sequences.
    args = SFTConfig(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        # num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 60,
        learning_rate = 2e-4,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.001,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Use TrackIO/WandB etc
    ),
)
```

```python
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

```python
trainer_stats = trainer.train()
```

```python
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(
    f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training."
)
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
```
